# Research Question

In major U.S. metropolitan areas, how do inflation-adjusted median salaries correlate with year-over-year job growth rates in computer and mathematical occupations from 2000 through 2023?

## Background and Prior Work

Over the past 20 years, the technology industry has grown fast, which create more and more job opportunities for people in computer and mathematical related fields. The yearly reports given by U.S. Bureau of Labor Statistics (BLS) about wages and job trends shows how demand and salaries change over time.  

Due to the technology advancements in the field, such as artificial intelligence and cloud computing, many industries need these skills for their businesses, therefore offering more job opportunities. However, higher job demand does not always mean higher salaries. Factors like automation, outsourcing, and economic problems can affect wages, keeping them the same or even making them lower, even when there are more jobs.  

Some studies have explored how job trends and salary growth are connected in computer and math related fields.  

Lockard and Wolf (2012) studied job predictions and found that computer and math-related jobs were expected to grow by 22% from 2010 to 2020. But they also found that wages do not always increase at the same speed. Automation, outsourcing, and economic changes affect salaries.<sup><a href="#fn1">1</a></sup> This supports the idea that job demand does not always lead to higher pay.  

A Bureau of Labor Statistics (2017) report showed that STEM jobs had higher-than-average wage growth. However, not all fields were the same. Computer-related jobs increased, but wages depended on global economic conditions, policy changes, and industry shifts.<sup><a href="#fn2">2</a></sup> This matches our research question, which shows how outside factors influence salaries in technology fields.

# Ethics & Privacy

Our study looks at how inflation-adjusted median salaries relate to employment rates in computer and mathematical occupations across major U.S. metropolitan areas from 2000 to 2023, using data from the **Bureau of Labor Statistics (BLS)**. While BLS provides reliable data, we recognize biases and ethical concerns that could impact our analysis.  

**Potential Bias in Our Research**  
- Our study focuses on median salaries and employment rates, but it doesn't consider the factor in wage differences by gender, race, or cost of living variations. Additionally, it doesn’t capture key parts of other compensation by the tech company like bonuses, stock options, and health benefits, which can make a big difference in the overall pay.

**Potential Bias in the Dataset**  
- The Bureau of Labor Statistics dataset relies heavily on employer-reported data, which doesn’t include freelancers or self-employed professionals in computer and math-related jobs. Because of this, our analysis might not fully capture employment trends for those roles. Plus, the dataset only presents aggregated statistics, it may miss wage disparities affecting underrepresented groups like women and minorities in tech.

**Confidentiality & Data Privacy**  
- BLS doesn’t just rely on surveys, it collects data through in-person visits, phone calls, and video interviews to improve accuracy and reduce job misclassification.  
- To protect privacy, BLS follows strict confidentiality policies, ensuring individual employers and respondents cannot be identified. Every dataset goes through a screening process before publication, making the data safe for public use and fully anonymized.  

**Post-Analysis Interpretation & Ethical Reporting**  
- There’s a risk that other reviewers may misinterpret our findings by assuming salary and job growth trends apply equally to all tech professionals, when in reality, disparities exist. To prevent this, we need to clearly communicate the dataset’s limitations and biases so that our conclusions aren’t misleading.  

**How We Address These Issues?**  
- We cross-check our findings with other sources like Glassdoor, Payscale, and LinkedIn to get a better understanding of freelancer and contract work trends.
- We look for wage outliers that could point to selection bias in employer-reported data.
- We need to clearly state dataset limitations, especially on the fact that it doesn’t include freelancers or other compensation benefits.  
- We avoid misleading charts or language that could overstate the link between salaries and job growth.  

By taking these steps, we ensure our research is transparent, ethical, and reflective of real-world job trends in the tech industry.
