# Research Question

What is the relationship between the **location quotient (LQ)** and **wage growth** in **Computer and Mathematical Occupations** across U.S. states from **2000 to 2023**? Specifically, does a higher or lower concentration of these occupations in a state (as measured by LQ) correspond to higher or lower wage growth over time?  

<u>**Key Definitions:**</u>  
- **Location Quotient (LQ):** A measure of occupational concentration in a state relative to the national average. An LQ **greater than 1** means the occupation is more concentrated in that state than the national average, while an LQ **less than 1** means it is less prevalent.  
- **Computer and Mathematical Occupations:** This category includes professions such as Computer Network Architects, Database Administrators, and Statisticians, among others.  
- **Study Period:** 2000 to 2023.  
- **State Groupings:**  
  - Republican-leaning states: Texas, Wyoming, Florida, Tennessee, Virginia.  
  - Democrat-leaning states: California, New York, Washington, Hawaii, Massachusetts.  

By analyzing trends in these states, we aim to determine whether a relationship exists between the concentration of computer and mathematical professionals and wage growth in these occupations.

## Background and Prior Work

Over the past 20 years, the technology jobs have grown rapidly, but we don't fully understand how wages in these jobs have changed across states. We are interested in examining how the increasing concentration of these jobs in the U.S. affects wages. To measure this relationship more concretely, we want to analyze if the concentration of Computer and Mathematical Occupations in a state (measured by LQ) affects how much wages grow over time.

Previous studies have explored the relationship between occupational concentration and wages, which provide useful insights for our research.

Wohl’s paper<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) refines the traditional location quotient method by incorporating what he calls the "method of reflections," which accounts not only for the concentration of occupations in a city but also for how widespread these occupations are across different cities. His findings reveal that cities with a more varied mix of jobs tend to experience greater wage growth, while those occupations that are common everywhere show smaller wage gains or even declines. This refined approach provides a richer understanding of how occupational diversity and prevalence interact with wage trends, offering a compelling perspective that resonates with our study on Computer and Mathematical Occupations. Although Wohl’s analysis spans all occupations in U.S. cities from 2006 to 2017, his methodology inspires us to consider additional dimensions beyond a simple concentration measure in our exploration of wage dynamics across states from 2000 to 2023.

Similarly, Echeverri‐Carroll and Ayala’s research<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) demonstrates that high‐tech cities command a wage premium—workers in these areas earn about 4.6% more than those in less tech‐intensive cities, even after accounting for individual skills. Their study attributes this premium to the positive effects of knowledge sharing , where a higher concentration of high‐tech industries enhances worker productivity. While their focus is broader, examining overall wage differences in high‐tech cities using 2000 Census data, the core idea—Their study shows that the way specialized industries are clustered in certain areas can affect wages—provides valuable context for our work. By applying similar principles to the specific case of Computer and Mathematical Occupations across U.S. states, our research builds on these insights to further explore the relationship between occupational concentration and wage growth.

While their research focuses on overall wage differences in high-tech cities using 2000 Census data, the core insight—that specialized industry clusters impact wages—guides our approach.

However, these studies did not explicitly analyze the correlation between LQ and wage growth over time in Computer and Mathematical Occupations. Despite these insights, gaps remain in the literature. While existing studies confirm that tech hubs experience wage growth, it remains unclear whether this trend extends to states with emerging tech sectors or lower occupational concentrations. Additionally, previous studies have not systematically compared how LQ influences wage growth in Republican-leaning versus Democrat-leaning states. By addressing these gaps, our study will contribute to a deeper understanding of how occupational concentration affects wage dynamics in the U.S.

### Data Manipulation and Imputation

For our analysis, we aim to examine the relationship between **location quotient (LQ)** and **wages** across different years. Therefore, our focus will be on analyzing trends at the **state level** rather than individual occupations. To achieve this, we will compute the **average values** of relevant variables for each **state and year**. Additionally, we will include a few other key columns that may provide valuable insights for further analysis.  

------------------
**Handling Missing LQ Data (2000-2009)**

A major issue in our dataset is the absence of **LQ (Location Quotient) values** for the years **2000 to 2009**. To address this, we will impute the missing values using the following approach:  

1. For each state, compute the average LQ from 2010 to 2023.  
2. Assign this average LQ to all missing values in 2000-2009 for the corresponding state. 

````python
mean_columns = ["TOT_EMP", "H_MEAN", "A_MEAN", "H_MEDIAN", "A_MEDIAN"]
mean_columns_w_lq = ["TOT_EMP", "H_MEAN", "A_MEAN", "H_MEDIAN", "A_MEDIAN", "LQ"]
state_means = {}
state_means_w_lq = {}

for year,df in new_dfs.items():       
    if (year>=2000 and year <=2009):
        df[mean_columns] = df[mean_columns].astype(float)
        state_means[year] = df.groupby("STATE")[mean_columns].mean().reset_index()
    else:
        df[mean_columns_w_lq] = df[mean_columns_w_lq].astype(float)
        state_means_w_lq[year] = df.groupby("STATE")[mean_columns_w_lq].mean().reset_index()


# Impute LQ values for 2000-2009 using these means
state_lq_means = (
    pd.concat(state_means_w_lq.values())  
    .groupby("STATE")["LQ"]  
    .mean()  
)
for year, df in state_means.items(): 
    df["LQ"] = df["STATE"].map(state_lq_means)  


combined_means = []
for year, df in state_means.items():
    df['YEAR'] = year
    df = df[['STATE', 'YEAR'] + mean_columns_w_lq]
    combined_means.append(df)
for year, df in state_means_w_lq.items():
    df['YEAR'] = year
    df = df[['STATE', 'YEAR'] + mean_columns_w_lq]
    combined_means.append(df)

lq_wage_data = pd.concat(combined_means, ignore_index=True)

# Sort the DataFrame by 'STATE' and then by 'YEAR'
lq_wage_data = lq_wage_data.sort_values(by=['STATE', 'YEAR']).reset_index(drop=True) 
lq_wage_data
````

````python
# Assign political affiliation to each state
republican_states = ["Tennessee", "Virginia", "Texas", "Wyoming", "Florida"]
democratic_states = ["California", "Massachusetts", "New York", "Washington", "Hawaii"]
lq_wage_data['Political_Affiliation'] = lq_wage_data['STATE'].apply(
    lambda x: 'Republican' if x in republican_states else 'Democrat'
)
lq_wage_data.head(1)
````

### Section 2: Understanding Wage Growth Through LQ and State-Level Analysis

#### High-LQ States vs. Low-LQ States

The **LQ** is used to assess whether the occupations of our interest are more or less concentrated in a given state compared to the national average.

To analyze wage trends, we categorie states into two groups based on their median LQ:

- High-LQ States (Median LQ > 1)
- Low-LQ States (Median LQ ≤ 1)

Median LQ > 1 means the state has consistently higher-than-average concentration across years, which makes sense for defining High-LQ states. The threshold of 1 is significant because it separates states with higher than average concentration from those with lower.

````python
median_lq = lq_wage_data.groupby('STATE')['LQ'].median()
lq_wage_data['Median_LQ'] = lq_wage_data['STATE'].map(median_lq)
lq_wage_data['LQ_Category'] = np.where(lq_wage_data['Median_LQ'] > 1, 'High-LQ', 'Low-LQ')

print("High-LQ States: ",lq_wage_data[lq_wage_data['LQ_Category'] == 'High-LQ']['STATE'].unique())
print("Low-LQ States: ",lq_wage_data[lq_wage_data['LQ_Category'] == 'Low-LQ']['STATE'].unique())
````

#### Comparing Wages Between High-LQ and Low-LQ States

Given that high-LQ states often have higher wages, we want to examine whether the LQ directly correlates with higher wage growth. Below, we examine the relationship between average LQ and annual mean wage per state to assess this pattern.

#### State-Level Wage Growth Rate Analysis

In this part, we will analyze the **growth rate of annual mean wages** across different states over time. The wage growth rate is calculated as the **percentage change in annual mean** wages from one year to the next, which helps us understand which states have experienced the highest wage growth over time.

Let's begin by calculating the annual wage growth rate for each state and visualizing the results.

````python

lq_wage_data = lq_wage_data.sort_values(by=["STATE", "YEAR"])

# Calculate the annual wage growth rate for each state
lq_wage_data['A_MEAN_growth_rate'] = lq_wage_data.groupby('STATE')['A_MEAN'].pct_change() * 100

# Drop rows with NaN values caused by the pct_change calculation (first year of each state's data)
growth_rate_data = lq_wage_data.dropna(subset=['A_MEAN_growth_rate'])

# Pivot data to format suitable for heatmap
heatmap_data = growth_rate_data.pivot(index='YEAR', columns='STATE', values='A_MEAN_growth_rate')

# Plot the heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, cmap='Reds', annot=False, fmt=".1f", linewidths=0.5, center=0)
plt.title("Annual Wage Growth Rate by State (Heatmap)")
plt.xlabel("States")
plt.ylabel("Years")
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()

plt.show()
````

While many states exhibit fluctuating wage growth rates over the years, no state shows a strictly consistent upward trend. However, some states display significantly higher volatility, meaning their wage growth rates shift more dramatically from year to year. The top three states with the highest fluctuations—as indicated by the frequent and extreme color changes in the heatmap—are **Wyoming**, **California**.

# Ethics & Privacy

Our study looks at the correlation between location quotient (LQ) and wage growth in **Computer and Mathematical Occupations** across U.S. states (5 democratics and 5 republicans) from 2000 to 2023, using data from the **Bureau of Labor Statistics (BLS)**. While BLS provides reliable data, we recognize biases and ethical concerns that could impact our analysis.  

**Potential Bias in Our Research**  
- Our study focuses on median salaries and employment rates, but it doesn't consider the factor in wage differences by gender, race, or cost of living variations. Additionally, it doesn’t capture key parts of other compensation by the tech company like bonuses, stock options, and health benefits, which can make a big difference in the overall pay.

**Potential Bias in the Dataset**  
- The Bureau of Labor Statistics dataset relies heavily on employer-reported data, which doesn’t include freelancers or self-employed professionals in computer and math-related jobs. Because of this, our analysis might not fully capture employment trends for those roles. Plus, the dataset only presents aggregated statistics, it may miss wage disparities affecting underrepresented groups like women and minorities in tech.

**Confidentiality & Data Privacy**  
- BLS doesn’t just rely on surveys, it collects data through in-person visits, phone calls, and video interviews to improve accuracy and reduce job misclassification.  
- To protect privacy, BLS follows strict confidentiality policies, ensuring individual employers and respondents cannot be identified. Every dataset goes through a screening process before publication, making the data safe for public use and fully anonymized.  

**Post-Analysis Interpretation & Ethical Reporting**  
- There’s a risk that other reviewers may misinterpret our findings by assuming salary and job growth trends apply equally to all tech professionals, when in reality, disparities exist. To prevent this, we need to clearly communicate the dataset’s limitations and biases so that our conclusions aren’t misleading.  

**How We Address These Issues?**  
- We cross-check our findings with other sources like Glassdoor, Payscale, and LinkedIn to get a better understanding of freelancer and contract work trends.
- We look for wage and location quotient outliers that could point to selection bias in employer-reported data.
- We need to clearly state dataset limitations, especially on the fact that it doesn’t include freelancers or other compensation benefits.  
- We avoid misleading charts or language that could overstate the link between salaries and job growth.  

By taking these steps, we ensure our research is transparent, ethical, and reflective of real-world job trends in the tech industry.
