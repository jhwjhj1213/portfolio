# Abstract
This project looks at how the concentration of Computer and Mathematical Occupations (measured by Location Quotient, LQ) relates to wage growth across U.S. states from 2000 to 2023. It explores whether states with more of these jobs see higher wage growth over time. The study also compares median wage growth between high-LQ and low-LQ states and between Democratic- and Republican-leaning states.

The analysis uses data from the Occupational Employment and Wage Statistics (OEWS) provided by the Bureau of Labor Statistics (BLS), focusing on ten states. The dataset was cleaned and standardized, including estimating missing LQ values for years before 2010. The study used exploratory data analysis (EDA), correlation tests, and non-parametric statistical methods to compare wage growth across different state categories.

Results show no strong relationship between LQ and wage growth, meaning a state’s concentration of tech jobs does not always lead to higher wages. However, high-LQ states tend to have more variation in wages, possibly due to differences in living costs. Additionally, Democratic-leaning states generally had higher wages than Republican-leaning states, likely due to industry differences and economic factors.

Overall, the results suggest that wage growth is influenced by multiple factors beyond occupational concentration. While this study does not directly examine all potential contributors, factors such as state policies, economic environments, and the presence of high-paying industries may play a role. These findings highlight the complexity of wage distribution patterns and offer a foundation for future research on wage trends across the U.S.

## Background and Prior Work

Over the past 20 years, the technology jobs have grown rapidly, but we don't fully understand how wages in these jobs have changed across states. We are interested in examining how the increasing concentration of these jobs in the U.S. affects wages. To measure this relationship more concretely, we want to analyze if the concentration of Computer and Mathematical Occupations in a state (measured by LQ) affects how much wages grow over time.

Previous studies have explored the relationship between occupational concentration and wages, which provide useful insights for our research.

Wohl’s paper<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) refines the traditional location quotient method by incorporating what he calls the "method of reflections," which accounts not only for the concentration of occupations in a city but also for how widespread these occupations are across different cities. His findings reveal that cities with a more varied mix of jobs tend to experience greater wage growth, while those occupations that are common everywhere show smaller wage gains or even declines. This refined approach provides a richer understanding of how occupational diversity and prevalence interact with wage trends, offering a compelling perspective that resonates with our study on Computer and Mathematical Occupations. Although Wohl’s analysis spans all occupations in U.S. cities from 2006 to 2017, his methodology inspires us to consider additional dimensions beyond a simple concentration measure in our exploration of wage dynamics across states from 2000 to 2023.

Similarly, Echeverri‐Carroll and Ayala’s research<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) demonstrates that high‐tech cities command a wage premium—workers in these areas earn about 4.6% more than those in less tech‐intensive cities, even after accounting for individual skills. Their study attributes this premium to the positive effects of knowledge sharing , where a higher concentration of high‐tech industries enhances worker productivity. While their focus is broader, examining overall wage differences in high‐tech cities using 2000 Census data, the core idea—Their study shows that the way specialized industries are clustered in certain areas can affect wages—provides valuable context for our work. By applying similar principles to the specific case of Computer and Mathematical Occupations across U.S. states, our research builds on these insights to further explore the relationship between occupational concentration and wage growth.

While their research focuses on overall wage differences in high-tech cities using 2000 Census data, the core insight—that specialized industry clusters impact wages—guides our approach.

However, these studies did not explicitly analyze the correlation between LQ and wage growth over time in Computer and Mathematical Occupations. Despite these insights, gaps remain in the literature. While existing studies confirm that tech hubs experience wage growth, it remains unclear whether this trend extends to states with emerging tech sectors or lower occupational concentrations. Additionally, previous studies have not systematically compared how LQ influences wage growth in Republican-leaning versus Democrat-leaning states. By addressing these gaps, our study will contribute to a deeper understanding of how occupational concentration affects wage dynamics in the U.S.

# Hypotheses


1. **Hypothesis for LQ vs. Wage Growth**
   
This test examines whether states with a higher concentration of tech jobs (LQ) saw greater wage growth over time (from 2000 to 2023).

**Null Hypothesis (H₀)**:
There is **no relationship** between the location quotient (LQ) in 2000 and wage growth between 2000 and 2023. LQ does not significantly influence how much wages increased over time.

$$
H_0: \text{LQ}_{2000} \quad \text{and} \quad \text{Wage Growth}_{2000-2023} \quad \text{are not correlated.}
$$

**Alternative Hypothesis (H₁)**:
There is a **positive relationship** between the location quotient (LQ) in 2000 and wage growth from 2000 to 2023. States with a higher concentration of tech jobs in 2000 (higher LQ) experienced higher wage growth over time.

$$
H_1: \text{LQ}_{2000} \quad \text{and} \quad \text{Wage Growth}_{2000-2023} \quad \text{are positively correlated.}
$$

---

2. **Hypothesis for LQ Category (High vs. Low) vs. Wage Growth**

This test examines whether there is a difference in **median wage growth** over time (from 2000 to 2023) between states with a **high LQ** and **low LQ**.

**Null Hypothesis (H₀)**:
There is **no significant difference** in median wage growth between states with high LQ and low LQ from 2000 to 2023.

$$
H_0: \text{Median Wage Growth}_{\text{High LQ}} = \text{Median Wage Growth}_{\text{Low LQ}}
$$

**Alternative Hypothesis (H₁)**:
States with **high LQ** experienced **greater median wage growth** over time (from 2000 to 2023) compared to states with **low LQ**.

$$
H_1: \text{Median Wage Growth}_{\text{High LQ}} \neq \text{Median Wage Growth}_{\text{Low LQ}}
$$

---

3. **Hypothesis for Political Affiliation (Democratic vs. Republican) vs. Wage Growth**

This test examines whether there is a difference in **median wage growth** over time (from 2000 to 2023) between states that lean **Democratic** and those that lean **Republican**.

**Null Hypothesis (H₀)**:
There is **no significant difference** in median wage growth between Democratic and Republican states from 2000 to 2023.

$$
H_0: \text{Median Wage Growth}_{\text{Democratic}} = \text{Median Wage Growth}_{\text{Republican}}
$$

**Alternative Hypothesis (H₁)**:
States that are **Democratic** experienced **greater median wage growth** over time (from 2000 to 2023) compared to states that are **Republican**.

$$
H_1: \text{Median Wage Growth}_{\text{Democratic}} \neq \text{Median Wage Growth}_{\text{Republican}}
$$

#### Data Manipulation and Imputation

For our analysis, we aim to examine the relationship between **location quotient (LQ)** and **wages** across different years. Therefore, our focus will be on analyzing trends at the **state level** rather than individual occupations. To achieve this, we will compute the **average values** of relevant variables for each **state and year**. Additionally, we will include a few other key columns that may provide valuable insights for further analysis.  

------------------
**Handling Missing LQ Data (2000-2009)**

A major issue in our dataset is the absence of **LQ (Location Quotient) values** for the years **2000 to 2009**. To address this, we will impute the missing values using the following approach:  

1. For each state, compute the average LQ from 2010 to 2023.  
2. Assign this average LQ to all missing values in 2000-2009 for the corresponding state. 

````python
mean_columns = ["TOT_EMP", "H_MEAN", "A_MEAN", "H_MEDIAN", "A_MEDIAN"]
mean_columns_w_lq = ["TOT_EMP", "H_MEAN", "A_MEAN", "H_MEDIAN", "A_MEDIAN", "LQ"]
state_means = {}
state_means_w_lq = {}

for year,df in new_dfs.items():       
    if (year>=2000 and year <=2009):
        df[mean_columns] = df[mean_columns].astype(float)
        state_means[year] = df.groupby("STATE")[mean_columns].mean().reset_index()
    else:
        df[mean_columns_w_lq] = df[mean_columns_w_lq].astype(float)
        state_means_w_lq[year] = df.groupby("STATE")[mean_columns_w_lq].mean().reset_index()


# Impute LQ values for 2000-2009 using these means
state_lq_means = (
    pd.concat(state_means_w_lq.values())  
    .groupby("STATE")["LQ"]  
    .mean()  
)
for year, df in state_means.items(): 
    df["LQ"] = df["STATE"].map(state_lq_means)  


combined_means = []
for year, df in state_means.items():
    df['YEAR'] = year
    df = df[['STATE', 'YEAR'] + mean_columns_w_lq]
    combined_means.append(df)
for year, df in state_means_w_lq.items():
    df['YEAR'] = year
    df = df[['STATE', 'YEAR'] + mean_columns_w_lq]
    combined_means.append(df)

lq_wage_data = pd.concat(combined_means, ignore_index=True)

# Sort the DataFrame by 'STATE' and then by 'YEAR'
lq_wage_data = lq_wage_data.sort_values(by=['STATE', 'YEAR']).reset_index(drop=True) 
lq_wage_data
````

````python
# Assign political affiliation to each state
republican_states = ["Tennessee", "Virginia", "Texas", "Wyoming", "Florida"]
democratic_states = ["California", "Massachusetts", "New York", "Washington", "Hawaii"]
lq_wage_data['Political_Affiliation'] = lq_wage_data['STATE'].apply(
    lambda x: 'Republican' if x in republican_states else 'Democrat'
)
lq_wage_data.head(5)
````

#### IQR of Annual Mean Wage vs. Median Wage Over Time

````python
# Calculate the IQR for each year
iqr_data = lq_wage_data.groupby("YEAR")["A_MEAN"].apply(lambda x: x.quantile(0.75) - x.quantile(0.25)).reset_index()
iqr_data.columns = ['YEAR', 'IQR_A_MEAN']

iqr_median_data = lq_wage_data.groupby("YEAR")["A_MEDIAN"].apply(lambda x: x.quantile(0.75) - x.quantile(0.25)).reset_index()
iqr_median_data.columns = ['YEAR', 'IQR_A_MEDIAN']
````

### Analysis 1: Correlation Testing: LQ vs. Wage Growth

We would like to answer the question whether high-LQ states saw more wage growth over the period 2000-2023. Let's first compute the wage growth.

**Wage Growth**

Wage growth can be measured as the percentage change in wages over time using the following formulas:

Hourly Mean Wage Growth:
$
\frac{H\_MEAN_t - H\_MEAN_{t0}}{H\_MEAN_{t0}}
$

Hourly Median Wage Growth:
$
\frac{H\_MEDIAN_t - H\_MEDIAN_{t0}}{H\_MEDIAN_{t0}}
$

Annual Mean Wage Growth:
$
\frac{A\_MEAN_t - A\_MEAN_{t0}}{A\_MEAN_{t0}}
$

Annual Median Wage Growth:
$
\frac{A\_MEDIAN_t - A\_MEDIAN_{t0}}{A\_MEDIAN_{t0}}
$

where:  
- $ t_0 $ is the starting year (e.g., 2000)  
- $ t $ is the ending year (e.g., 2023)  


````python
# Filter data for the years 2000 and 2023
df_2000 = lq_wage_data[lq_wage_data['YEAR'] == 2000][['STATE', 'H_MEAN', 'H_MEDIAN', 'A_MEAN', 'A_MEDIAN', 'LQ','LQ_Category','Political_Affiliation']]
df_2023 = lq_wage_data[lq_wage_data['YEAR'] == 2023][['STATE', 'H_MEAN', 'H_MEDIAN', 'A_MEAN', 'A_MEDIAN','LQ']]

# Rename columns
df_2000 = df_2000.rename(columns={'H_MEAN': 'H_MEAN_2000', 'H_MEDIAN': 'H_MEDIAN_2000',
                                  'A_MEAN': 'A_MEAN_2000', 'A_MEDIAN': 'A_MEDIAN_2000', 'LQ': 'LQ_2000'})
df_2023 = df_2023.rename(columns={'H_MEAN': 'H_MEAN_2023', 'H_MEDIAN': 'H_MEDIAN_2023',
                                  'A_MEAN': 'A_MEAN_2023', 'A_MEDIAN': 'A_MEDIAN_2023', 'LQ': 'LQ_2023'})
# Merge the data on STATE
df_growth = df_2000.merge(df_2023, on="STATE")

# Calculate wage growth
df_growth['H_MEAN_Growth'] = (df_growth['H_MEAN_2023'] - df_growth['H_MEAN_2000']) / df_growth['H_MEAN_2000']
df_growth['H_MEDIAN_Growth'] = (df_growth['H_MEDIAN_2023'] - df_growth['H_MEDIAN_2000']) / df_growth['H_MEDIAN_2000']
df_growth['A_MEAN_Growth'] = (df_growth['A_MEAN_2023'] - df_growth['A_MEAN_2000']) / df_growth['A_MEAN_2000']
df_growth['A_MEDIAN_Growth'] = (df_growth['A_MEDIAN_2023'] - df_growth['A_MEDIAN_2000']) / df_growth['A_MEDIAN_2000']

df_growth[['STATE','H_MEAN_Growth','H_MEDIAN_Growth','A_MEAN_Growth','A_MEDIAN_Growth']].head()
````

We have only 10 states and thus 10 data points, which is a very small sample size. We will use a non-parametric correlation test, such as Spearman's rank correlation, which doesn't assume normality or linearity. It looks at the monotonic relationship between variables rather than the specific form of the relationship.

````python
plt.figure(figsize=(5, 4))
sns.scatterplot(x=df_growth['LQ_2000'], y=df_growth['A_MEDIAN_Growth'])
plt.title("LQ (2000) vs. Annual Median Wage Growth (2000-2023)")
plt.xlabel("LQ (2000)")
plt.ylabel("Hourly Median Wage Growth (2000-2023)")
plt.show()
````

````python
def hypo_test(p, cutoff):
    if p < cutoff:
        print(f"The correlation is statistically significant (p < {cutoff}).")
    else:
        print(f"The correlation is NOT statistically significant (p > {cutoff}).")
````

````python
from scipy.stats import spearmanr

corr, p_value = spearmanr(df_growth['LQ_2000'], df_growth['A_MEDIAN_Growth'])

print(f"Spearman Rank Correlation between LQ (2000) and Annual Median Wage Growth (2000-2023):")
print(f"Correlation Coefficient: rho = {corr:.4f}")
print(f"P-value: p = {p_value:.4f}")
hypo_test(p_value, 0.05)
````

### Analysis 2: Difference in Wage Growth between High-LQ states and Low-LQ states


We would like to assess whether there is a significant difference in the **median wage growth** between **high LQ states** and **low LQ states**.

Given that we have a small dataset, with only 5 data points for states with high LQ and 5 data points for states with low LQ, we need to choose an appropriate statistical test. 

The **Mann-Whitney U Test** is a non-parametric test that is particularly well-suited for small sample sizes and when the assumptions required for parametric tests (such as t-tests) may not be met. This test compares the ranks of values between two independent groups, making it ideal for our case where normality cannot be assumed due to the limited sample size.



````python
# Plotting a boxplot to visualize the distribution of Annual Median Wage Growth 
# between High-LQ and Low-LQ
plt.figure(figsize=(5, 4))
sns.boxplot(x='LQ_Category', y='A_MEDIAN_Growth', data=df_growth, hue='LQ_Category')
plt.title("Annual Median Wage Growth Across High_LQ and Low_LQ States (2000-2023)")
plt.ylabel("Annual Median Wage Growth (2000-2023)")
plt.show()
````

````python
from scipy.stats import mannwhitneyu

high_lq_growth = df_growth[df_growth['LQ_Category'] == 'High-LQ']['A_MEDIAN_Growth']
low_lq_growth = df_growth[df_growth['LQ_Category'] == 'Low-LQ']['A_MEDIAN_Growth']

u_stat, p_value_u = mannwhitneyu(high_lq_growth, low_lq_growth)


print(f"Mann-Whitney U Test between High_LQ and Low_LQ wage growth:")
print(f"U-statistic: {u_stat:.4f}")
print(f"P-value: {p_value_u:.4f}")
hypo_test(p_value_u, 0.05)
````
